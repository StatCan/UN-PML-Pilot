```python
import pandas as pd
```

```python
time_res = pd.read_csv("time.log", sep=";")
acc_res = pd.read_csv("accuracy.log", sep=";")

merged_results = time_res.merge(acc_res, how='left', on=['Rounds', 'Epochs'])
```

```python
merged_results
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

rounds_5 = merged_results[merged_results['Rounds'] == 5]

ax.plot(rounds_5['Epochs'], rounds_5['Time'], rounds_5['Accuracy'])
```
